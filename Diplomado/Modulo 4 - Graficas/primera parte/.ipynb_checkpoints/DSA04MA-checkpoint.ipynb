```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Para la inclusión de los siguientes recursos, **asegúrate de actualizar la ruta**

```python
countries = pd.read_csv("world_table_country.csv")
languages = pd.read_csv("world_table_countrylanguage.csv")
```

```python
sns.set_theme(style="darkgrid")
```

# **TEMA 1. Subplots**

```python
centralAmerica = countries.loc[(countries['Continent']=='North America') & (countries['Region']=='Central America'), ['Code', 'Name','SurfaceArea', 'Population', 'LifeExpectancy', 'GNP']]
centralAmerica.set_index('Code', inplace=True)
centralAmerica
```

***Subplots con matplotlib***

```python
fig, axs = plt.subplots(2,2,figsize=(10,8))
axs[0,1].plot(centralAmerica['SurfaceArea'])
axs[0,1].set_title('Superficie territorial')
axs[0,1].set_ylabel('Km cuadrados')
```

***Subplots con Pandas***

```python
fig, axs = plt.subplots(2,2, figsize=(15,10))
plt.subplots_adjust(wspace=0.2, hspace=0.32)
centralAmerica.plot(x='Name', y='SurfaceArea', title='Superficie territorial', ax=axs[0,1], rot=90)
```

```python
fig, axs = centralAmerica.plot(subplots=True, layout=(2,2), figsize=(15,10), rot=0)
```

***Subplots con Seaborn***

```python
fig, axs = plt.subplots(2,2,figsize=(15,10))
plt.subplots_adjust(wspace=0.2, hspace=0.32)
sns.lineplot(x=centralAmerica['Name'], y=centralAmerica['SurfaceArea'], ax=axs[0,1]);
axs[0,1].tick_params(axis='x', rotation=90)
```

# **TEMA 2. Anotaciones**

## 2.1 Texto en coordenadas

```python
languagesMEX = languages.loc[languages['CountryCode'] == 'MEX', ['Language','Percentage']]
languagesMEX.set_index('Language', inplace=True)
languagesMEX
```

```python
plt.plot(languagesMEX)
plt.text(3,92.1,'92.1%')
```

```python
plt.plot(languagesMEX)
plt.text(3,92.1,'92.1%', family='serif', fontsize=18, fontweight='bold', style='italic', color='red', horizontalalignment='center')
```

```python
plt.plot(languagesMEX)
plt.text(3,92.1,'92.1%', bbox=dict(facecolor='red', alpha=0.3, pad=5))
```

```python
plt.plot(languagesMEX)
for i in range(languagesMEX.shape[0]):
  plt.text(i,languagesMEX.iloc[i,0],str(languagesMEX.iloc[i,0])+'%')
```

## 2.2 Trazado de tablas

***Tablas con matplotlib***

```python
plt.plot(languagesMEX)
plt.tick_params(labelbottom=False)
languagesMEX_T = languagesMEX.T
plt.table(cellText=languagesMEX_T.values, colLabels=languagesMEX_T.columns)
```

***Tablas con Pandas***

```python
languagesMEX.plot(table=True, xticks=[], xlabel='')
```
