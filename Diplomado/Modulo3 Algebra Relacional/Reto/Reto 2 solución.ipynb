**El tiempo de servicio promedio de los trabajadores**

```
print('Tiempo de servicio promedio de los trabajadores: ', empleados['Time_of_service'].mean())
```

**El porcentaje de hombres y mujeres en el archivo**

```
porcentaje_genero = empleados[['Employee_ID', 'Gender']].groupby('Gender').count()
porcentaje_genero['Porcentaje'] = porcentaje_genero['Employee_ID'] / porcentaje_genero['Employee_ID'].sum() * 100
porcentaje_genero
```

**El tiempo máximo sin promoción de las personas que llevan más de 15 años de servicio**

```
personasMas15años = empleados.loc[empleados['Time_of_service']>15, 'Time_since_promotion']
personasMas15años.max()
```

**La edad promedio en la que ingresaron las personas a la empresa, genera una nueva columna derivada mediante la formula Age - Time_of_service**

```
empleados['edad_ingreso'] = empleados['Age'] - empleados['Time_of_service']
empleados
```

```
serieAux = empleados['Age'] - empleados['Time_of_service']
print('Prom con na= ', serieAux.mean())
serieAux = serieAux.dropna()
serieAux
print('Prom con na= ', serieAux.mean())
```

```
print('Edad promedio de ingreso a la empresa: ', empleados['edad_ingreso'].mean())
```

*Un dataframe con las columnas Employee_ID, Relationship, Decision_skill_process que sólo contenga los registros de mujeres, solteras y con compensación tipo 2 o 3**

```
solteras_comp_2_3 = empleados.loc[(empleados['Gender']=='F') & (empleados['Relationship_Status']=='Single') & ((empleados['Compensation_and_Benefits']=='type2') | (empleados['Compensation_and_Benefits']=='type3')), ['Employee_ID', 'Relationship_Status', 'Decision_skill_possess']]
solteras_comp_2_3
```

**Una columna derivada con el valor "on time" si cuando tuvo su última promoción tenía menos de 30 años. En caso contrario colocar el valor de "out of time"**

```
empleados['edad_ultima_promocion'] = empleados['Age'] - empleados['Time_since_promotion']
empleados['promotion_status'] = ["on time" if edad_ultima_promocion < 30 else "out of time" for edad_ultima_promocion in empleados['edad_ultima_promocion']]
empleados
```
