```python
#Importamos Pandas
import pandas as pd
```

```python
################
#Cargamos Datos - Test.csv, Train.csv, Test2.csv, Train2.csv
################
test = pd.read_csv("/content/drive/MyDrive/Retos/Reto3/Test.csv")
#test.info()
test.head(5)
```

```python
train = pd.read_csv("/content/drive/MyDrive/Retos/Reto3/Train.csv")
#train.info()
train.head(5)
```

```python
test2 = pd.read_csv("/content/drive/MyDrive/Retos/Reto3/Test2.csv")
test2.head(5)
```

```python
train2 = pd.read_csv("/content/drive/MyDrive/Retos/Reto3/Train2.csv")
train2.head(5)
```

```python
# Concatenamos la primera parte
data = [test, train]
dataConcat = pd.concat(data)
dataConcat
```

```python
# Concatenamos la segunda parte
data2 = [test2, train2]
dataConcat2 = pd.concat(data2)
#Se identificó un espacio en blanco en el nombre de las columnas, con rename se corrigió este problema 
dataConcat2=dataConcat2.rename(columns = {' Computer_OS':'Computer_OS', ' Computer_OS_version':'Computer_OS_version', ' Computer_tickets':'Computer_tickets', ' Mobile_OS':'Mobile_OS', ' Mobile_OS_version':'Mobile_OS_version', ' Mobile_tickets':'Mobile_tickets'})
#dataConcat2.info()
dataConcat2.head(5)
```

```python
#Juntamos los 2 dataframes en uno solo
empleados = dataConcat.merge(dataConcat2, how='left')
#Re-Index
empleados.reset_index(drop=True)
#Aqui ya tenemos todos los csv's contatenados
empleados
```

```python
#Se asigna el valor "No registrado" en los siguiente campos: Computer_OS, Computer_OS_version, Mobile_OS
empleados[['Computer_OS', 'Mobile_OS']] = empleados[['Computer_OS', 'Mobile_OS']].fillna('no registrado')
#empleados[empleados['Employee_ID'] == 'EID_2279']
print("Asignación de valor No registrado a los siguiente campos: Computer_OS, Computer_OS_version, Mobile_OS")
empleados
```

```python
#El porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora 
generaPorcentaje = empleados[['Employee_ID', 'Computer_OS']].groupby('Computer_OS').count()
generaPorcentajeC = generaPorcentaje['Employee_ID'] / generaPorcentaje['Employee_ID'].sum() * 100
print("Porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora ")
generaPorcentajeC
```

```python
#El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil 
generaPorcentaje = empleados[['Employee_ID', 'Mobile_OS']].groupby('Mobile_OS').count()
generaPorcentajeM = generaPorcentaje['Employee_ID'] / generaPorcentaje['Employee_ID'].sum() * 100
print("Porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo móvil")
generaPorcentajeM
```

```python
#La edad promedio de la gente que utiliza cada sistema operativo de escritorio
#Computer_OS
edadPromedioE = empleados[['Computer_OS', 'Age']].groupby("Computer_OS").mean()
edadPromedioE
```

```python
#La edad promedio de la gente que utiliza cada sistema operativo de Mobil
#Mobile_OS
edadPromedioM = empleados[['Mobile_OS', 'Age']].groupby("Mobile_OS").mean()
edadPromedioM
```

```python
#El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio  
#Education_Level
nivelEducativoE = empleados[['Computer_OS', 'Education_Level']].groupby("Computer_OS").mean()
nivelEducativoE
```

```python
#El nivel educativo promedio de la gente que utiliza cada sistema operativo de Mobil  
#Education_Level
nivelEducativoM = empleados[['Mobile_OS', 'Education_Level']].groupby("Mobile_OS").mean()
nivelEducativoM
```

```python
#El número total de tickets generados por la gente con cada sistema operativo de escritorio 
# Computer_tickets
empleados
TicketsE = empleados[['Computer_OS', 'Computer_tickets']].groupby("Computer_OS").mean()
TicketsE
```

```python
#El número total de tickets generados por la gente con cada sistema operativo de Mobile 
# Computer_tickets
empleados
TicketsM = empleados[['Mobile_OS', 'Computer_tickets']].groupby("Mobile_OS").mean()
TicketsM
```

```python
#Se envía el resultado de cada solicitud de información a un archivo XLS[X], donde el resultado de cada consulta debe ir en su propia hoja de trabajo 
with pd.ExcelWriter('reto3-OKSM.xlsx') as writer:  
    empleados.to_excel(writer, sheet_name='concatenadoGral')
    generaPorcentajeC.to_excel(writer, sheet_name='porcentajeCompu')
    generaPorcentajeM.to_excel(writer, sheet_name='porcentajeMovil')
    edadPromedioE.to_excel(writer, sheet_name='edadPromedioEscritorio')
    edadPromedioM.to_excel(writer, sheet_name='edadPromedioMobile')
    nivelEducativoE.to_excel(writer, sheet_name='nivelEducativoEscritorio')
    nivelEducativoM.to_excel(writer, sheet_name='nivelEducativoMobile')
    TicketsE.to_excel(writer, sheet_name='TicketsEscritorio')
    TicketsM.to_excel(writer, sheet_name='TicketsMobile')
```
