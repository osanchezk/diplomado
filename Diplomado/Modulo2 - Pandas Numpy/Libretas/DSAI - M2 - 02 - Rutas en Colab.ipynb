**Usando una ruta absoluta de google drive**

---

1. Colocar el archivo en nuestro drive (por ejemplo se coloca el archivo **"characters.csv"** en nuestro drive en la carpeta **Colab Notebooks/M2DCData/**)
2. Importar drive de **google.colab**
3. **Montar** el drive
4. Usar la **ruta completa** comenzando con la ruta base de nuestro drive que sería **/content/drive/MyDrive** y a partir de ahí la ruta donde se encuentre el archivo que desean utilizar (por ejemplo utilizamos **/content/drive/MyDrive/Colab Notebooks/M2DCData/characters.csv**)
5. Si es la primera vez que se utiliza el drive aparece una URL para generar una authorización, hay que entrar, permitir el acceso y copiar el código de autorización generado para pegarlo en el campo **"Enter your authorization code:"**
6. Si el código es correcto y la ruta con el archivo existen entonces el script se ejecutara y obtendran los resultados deseados accesando a un archivo en el drive por medio de su ruta absoluta.


```python
from google.colab import drive 
import pandas as pd
drive.mount('/content/drive')
personajesSW = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/M2DCData/characters.csv")
personajesSW.head(8)
```
